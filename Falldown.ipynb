```python
import uvage
import pygame
import random



#Parameters

camera = uvage.Camera(800,600)

print(camera.top, camera.bottom, camera.left, camera.right)

fallrate = 5
scrollrate = 7
terminalfallrate = 10

box_size_xy = 50
bars = 4
    
game_on = False
game_over = False
score = 0


ticker = 0

#Gameboxes

#Creates Evenly Distributed Bars
barslist = []
totalarea = camera.bottom

for i in range(1, bars + 1):
    left_or_right = random.randint(0,1)

    if left_or_right == 0:
        randombarsize_l = random.randint(1, camera.right/2)
        randombarsize_r = (camera.right - randombarsize_l) - 100

    elif left_or_right == 1:
        randombarsize_r = random.randint(1, camera.right/2)
        randombarsize_l = (camera.right - randombarsize_r) - 100
        

    bar_l = uvage.from_color(camera.left + randombarsize_l/2, totalarea, 'yellow', randombarsize_l, 25)
    bar_r = uvage.from_color(camera.right - randombarsize_r/2, totalarea, 'yellow', randombarsize_r, 25)
    barslist.append([bar_l, bar_r])
    totalarea -= 150

box = uvage.from_color(400, 100, "green", box_size_xy, box_size_xy)

#Creates A Box That Contains the Game
vertical_bounds = [uvage.from_color(0, camera.top, "black", 10000, 1), 
uvage.from_color(0, camera.bottom, "black", 10000 , 1)]

horizontal_bounds = [uvage.from_color(camera.right, 0, "black", 1, 10000),
uvage.from_color(camera.left, 0, "black", 1, 10000)]

#Sets the Speed for the Box
box.yspeed = 0
box.xspeed = 50

def tick():
    global game_on, game_over, score, barslist
    
    #Scrolls the Bars and Box Fall
    if game_on == True:
        score += 1
        box.move(0, box.yspeed)
        for bars in barslist:
            for bar in bars:
                bar.move(0, -scrollrate)

    if uvage.is_pressing("space"):
        game_on = True

    #Inputs

    if uvage.is_pressing("right arrow"):
        box.move(box.xspeed, 0)

    if uvage.is_pressing("left arrow"):
        box.move(-box.xspeed, 0)

    #Collision

    #If the Box is not on a bar, the fall accelerates
    for bars in barslist:
        for bar in bars:
            speedtickcounter = 0

            if box.touches(bar):
                box.yspeed = 0
                speedtickcounter = 0
            else:
                speedtickcounter += 1
                box.yspeed += fallrate * speedtickcounter
                if box.yspeed > terminalfallrate:
                    box.yspeed = terminalfallrate
                
            if bar.touches(box):
                box.move_to_stop_overlapping(bar)

    #Check if the Box is touching the bounds
    for bounds in vertical_bounds:
        if box.touches(bounds):
            box.yspeed = 0
    
    for bounds in horizontal_bounds:
        if box.touches(bounds):
            box.move_to_stop_overlapping(bounds)
        else:
            box.xspeed = 50

    if box.touches(vertical_bounds[0]): #Upper Bound
        game_on = False
        game_over = True

    #Spawning in Bars After it Touches the Top

    for bars in barslist:
        for bar in bars:
            if bar.touches(vertical_bounds[0]):
                left_or_right = random.randint(0,1)
                if left_or_right == 0:
                    randombarsize_l = random.randint(1, camera.right/2)
                    randombarsize_r = (camera.right - randombarsize_l) - 100
                elif left_or_right == 1:
                    randombarsize_r = random.randint(1, camera.right/2)
                    randombarsize_l = (camera.right - randombarsize_r) - 100
                    

                bar_l = uvage.from_color(camera.left + randombarsize_l/2, camera.bottom, 'yellow', randombarsize_l, 25)
                bar_r = uvage.from_color(camera.right - randombarsize_r/2, camera.bottom, 'yellow', randombarsize_r, 25)

                bars[0] = bar_l
                bars[1] = bar_r

    #Display
    camera.clear('white')

    for bars in barslist:
        print(bars)
        for bar in bars:
            camera.draw(bar)

    for bounds in vertical_bounds:
        camera.draw(bounds)

    for bounds in horizontal_bounds:
        camera.draw(bounds)

    camera.draw(box)
    camera.draw(uvage.from_text(400, 50, str(score), 50, "Red", bold=True))

    # Game Over
    if game_over == True:
        print('gameover')
        camera.draw(uvage.from_text(400, 400, 'Game Over', 100, "Red", bold=True))
        camera.draw(uvage.from_text(400, 200, 'Press r to Reset', 100, "Red", bold=True))
        box.xspeed = 0
        box.yspeed = 0
    
    if uvage.is_pressing("r"):
        game_on = True
        game_over = False
        camera.clear('white')
        box.center = [400, 100]
        box.yspeed = fallrate
        score = 0
    

    camera.display()   
    
    #if game_over == True:

#Timer

ticks_per_second = 30

uvage.timer_loop(ticks_per_second, tick)
```
